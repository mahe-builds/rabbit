# Bronze Table creation

## source_crm dataset

### cust_info dataset 

```python
%sql
CREATE TABLE my_catalog.bronze.cust_info
(
    cst_id INT NOT NULL,
    cst_key STRING,
    cst_firstname STRING,
    cst_lastname STRING,
    cst_marital_status STRING,
    cst_gndr STRING,
    cst_create_date DATE
)
USING DELTA
COMMENT 'Customer Information table';


```

```python
%sql
CREATE OR REPLACE TABLE my_catalog.bronze.erp_cust_info_raw
(
    cst_id INT,
    cst_key STRING,
    cst_firstname STRING,
    cst_lastname STRING,
    cst_marital_status STRING,
    cst_gndr STRING,
    cst_create_date DATE
)
USING DELTA
COMMENT 'Customer Information table -Bronze layer';
```

### prd_info dataset

```python
%sql
CREATE TABLE my_catalog.bronze.prd_info
(
    prd_id INT NOT NULL,
    prd_key STRING,
    prd_nm STRING,
    prd_cost INT,
    prd_line STRING,
    prd_start_dt DATE,
    prd_end_dt DATE
)
USING DELTA
COMMENT 'Product Information table - Bronze layer';
```

### sales_details dataset

```python
%sql
CREATE TABLE my_catalog.bronze.sales_details
(
    sls_ord_num STRING NOT NULL,
    sls_prd_key STRING,
    sls_cust_id INT,
    sls_order_dt DATE,
    sls_ship_dt DATE,
    sls_due_dt DATE,
    sls_sales DECIMAL(12,2),
    sls_quantity INT,
    sls_price DECIMAL(12,2)
)
USING DELTA
COMMENT 'Sales detail Information - Bronze layer';
```
